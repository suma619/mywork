```python
import csv
def extract1(i):
    
    data=''.join(i)    
    ex=data[data.index('\"'):data.index('\" \"')+1]
    return ex

def extract2(i):
    ex=i[i.index('\"'):i.index('HTTP/')]
    return ex

def extract3(i):
    ex=i[i.index('\" ')+1:i.index(' \"')]
    return ex

def extract4(i):
    ex=i[i.index(' ')+1:i.index(' \"')]
    ex=ex[ex.index('\" ')+1:]
    return ex
    
    
def countX(lst, x): 
    
    km=x
    count = 0
    x=x.replace(" ","")
    found=[]
    for ele in lst:
        ele=ele.replace(" ","")
        if(ele == x): 
            count = count + 1
        #print(count)
    #if(count>1):
    found.append(km)
            #print(found)
    #print(found)
    return count,found
    
    
def analyze(list):
    sucess_count=0
    failure_count=0
    forbidden_err=0
    not_found=0
    moved_permanently=0
    found1=0
    not_modified=0 
    for item in list:
        ex_4=extract4(item)
        code=ex_4[0:4]
        code=int(code)
        if(code==200):
        #print(code,"sucess")
            sucess_count=sucess_count+1
        else:
            failure_count=failure_count+1
            if (code==403):
                forbidden_err=forbidden_err+1
            elif (code==404):
                not_found=not_found+1
            elif (code==301):
                moved_permanently=moved_permanently+1
            elif (code==302):
                found1=found1+1
            else:
                not_modified=not_modified+1
        return sucess_count,failure_count,forbidden_err,not_found,moved_permanently,found1,not_modified
        
    

file="C:/Users/suma/Desktop/work/logs/access_log1"
#path="C:/Users/suma/Desktop/work/"
#file2="C:/Users/suma/Desktop/work/logs/access_log"
#file3="C:/Users/suma/Desktop/work/logs/access_log4"
file3="C:/Users/suma/Desktop/work/logs/access_log5"

import os
#name=os.path.basename("C:/Users/suma/Desktop/work/access_log")
count=0
fh=open(file,"r")
#fh2=open(file2,"w+")
fh3=open(file3,"w+")
for k in fh:
    exe=extract1(k)
    #print(exe)
    urle=extract2(exe)
    fh3.write(exe+"\n")
    #print(urle)
lineList = list()


lineList = [line.rstrip('\n') for line in open("logs/access_log5")]
print ("  Sucess  Failure Forbidden NOt_Found Mvd_Perm Found Not_Modified")
print("   200      any    403       404       301      302   304")
k=1
for i in lineList:
    
    coc,found=countX(lineList,i)
    #print(found)
    sucess,failure,forbidden,not_fo,moved_perma,fou,not_modi=analyze(found)
    print(k," ",sucess,"     ",failure,"     ",forbidden,"     ",not_fo,"     ",moved_perma,"     ",fou,"     ",not_modi)
    k=k+1

    row=[i,sucess,failure,forbidden,not_fo,moved_perma,fou,not_modi]
    with open('logs1.csv', 'a') as csvFile:

        writer = csv.writer(csvFile)
        writer.writerow(row)
        
    



    
```
