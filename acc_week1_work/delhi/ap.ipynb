```python
import pandas as pd
import getpass
from datetime import date

def sort_21(exp):
    
    exp=str(exp)
    sum=0
    
    bad_chars = ['m', 'o', 'n','t','h','s'] 
    for i in bad_chars : 
        exp = exp.replace(i, '')
    year=exp.split(" ")
    year_1=year[0]
    #print(year,exp)
    month=year[2]
    #sum=year_1
    #print(month)
    if month=="0":
        return(year_1)
    else:
        sum=round(int(year_1)+int(month)/12,2)
    #print(sum)
        return sum
    
def sort_ctc(exp):
    exp=str(exp)
    sum=0
    
    bad_chars = ['T', 'h', 'o','u','s','a','n','d','(',')','L','k'] 
    for i in bad_chars : 
        exp = exp.replace(i, '')
    #print(exp)
    ctc=exp.split(" ")
    #print(ctc)
    lakh=ctc[0]
    thou=ctc[2]
    #print(lakh,thou)
    if lakh=="0":
        #print("yess")
        return ("0")
    if thou==0:
        return lakh
    else:
        #print(thou)
        sum=lakh+"."+thou
    #print(sum)
        return sum

def sort_no(no):
    import numpy as np
    no=str(no)
    
    if len(no)==10:
        return (no)
    else:
        return(np.nan)
    
file="sales delhi data.. one lakh .xlsx"
file_name="data/sales delhi data.. one lakh .xlsx"
dfs = pd.read_excel(file_name, sheet_name="Sales Delhi")

df1=dfs

df1["experience"]= df1["experience"].replace("Years",'',regex=True) 
df1['experience']=df1['experience'].apply(sort_21)


df1['ctc']=df1['ctc'].apply(sort_ctc)

df1['phone']=df1['phone'].apply(sort_no)
df1['phone2'] = df1['phone2'].str.split('-').str[1]
df1['phone2']=df1['phone2'].apply(sort_no)

df1['filename']=file


today = date.today()

user_name=getpass.getuser()

created_by="Sumesh Shetty"
creation_date="2019-11-19"

df1['creation_date']=creation_date
df1['created_by']=created_by

df1['modified_by']=user_name
df1['modified_date']=today 

df1.to_excel("processed/"+file,index=False)
```
